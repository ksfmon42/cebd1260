```python
# Chunking and Chinking 
def process_content():
    try:
        for s in tokenized:
            words = nltk.word_tokenize(s)
            filtered = [w for w in words if not w in stop_words]
            tagged = nltk.pos_tag(filtered)
            #chunkGram = r"""Chunk:{<RB.?>*<VB.?><JJ>?<NN>?<NNP>?<NN>?}"""
            #chunkGram = r"""Chunk:{<.*>+}
            #                }<VB.?|IN|DT|TO>+{"""
            chunkGram = r"""Chunk:{<JJ>?<NN>*<NNP>*}"""
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            print(chunked)
            chunked.draw()
    except Exception as e:
        print(str(e))
```

```python
process_content()
```

```python
# Name Entity Recognition
def process_content2():
    try:
        for s in tokenized:
            words = nltk.word_tokenize(s)
            filtered = [w for w in words if not w in stop_words]
            tagged = nltk.pos_tag(filtered)
            namedEnt = nltk.ne_chunk(tagged, binary=True) # Binary=True to remove categorization of the entity
            print(namedEnt)
            namedEnt.draw()
    except Exception as e:
        print(str(e))
```

```python
process_content2()
```
